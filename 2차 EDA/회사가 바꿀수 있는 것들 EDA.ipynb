# 진행방향소개

- 1. 자발적 이직을 막는 것
>회사가 통제할수 있는 변인으로 해결 할 수 있다.

- 2. turnover 할 수 있는 사람을 뽑지 않는 것 
>사람을 뽑기 전에 알 수 있는 데이터(교육수준 등)으로 알 수 있을 거 같다. 
>윤리적인 문제가 있으므로, 부서 별, 성별, 연령 별 특징 정도로 해두자.

- control 할 수 있는것
>출장 빈도, 월급, 스톡옵션, 부서, 직무수준, 직무역할 등

- control 할 수 없는것
>교육 수준, 교육 분야, 등

- 업무 환경 만족도, 직무 만족도, 직무몰입 
>이직과 관련이 굉장히 큰 Columns들이다

>그러나 이 Columns들은 회사가 직접적으로 control하기 힘들고, 다른 변인들에 의해 좌우되는 특성이 있다.

- 이 데이터로 회사에게 어떻게 유의미한 인사이트를 줄 수 있을까
>하나의 그룹을 정하고(회사에 유의미하고 중요한 그룹)+ 그룹과 다른 그룹의 특성을 비교하는데, 회사가 관리할 수 있는 변수에 한해서 비교한다면 그들의 이직을 어느 정도 막는데 기여할 수 있겠다.

>핵심인재의 자발적 이직 관리가 회사의 경영진 그룹에서 가장 중요함. 

>핵심인재의 특징 : 다른 그룹과 비교

- 핵심인재를 어떻게 정의 할 수 있을까?
>1.임금상승률이 높은 사람들로 간접적으로 알 수 있을것이라 생각.
>2.돈많이 받는 사람.
>3.어린 사람.

- 사전 검토에서 이 회사는 특히 어린 사람의 이직이 심한 문제로 나타났다. 
- 따라서 어리고, 임금상승률이 높은 사람을 한 그룹으로 선정하여 나머지 그룹과 비교하고자 한다.
- 여러 변수들이 정수형으로 되어 있기 때문에 효율적인 시각화를 위해서 그룹을 나눌 예정

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns

import warnings

warnings.filterwarnings('ignore')

data = pd.read_csv('../data/WA_Fn-UseC_-HR-Employee-Attrition.csv')
data.head(5)
```

# Feature Engineering 

## 연속형 > 범주화

```python
data['Rate_band'] = pd.qcut(data['MonthlyRate'], q=4, labels=[1,2,3,4])
data['Income_band'] = pd.qcut(data['MonthlyIncome'], q=4,labels=[1,2,3,4])
data['SalaryHike_band'] = pd.qcut(data['PercentSalaryHike'], q=4,labels=[1,2,3,4])
data['HomeDistance_band'] = pd.qcut(data['DistanceFromHome'], q=4, labels=[1,2,3,4])

data['Rate_band']=data['Rate_band'].astype('int')
data['Income_band']=data['Income_band'].astype('int')
data['SalaryHike_band']=data['SalaryHike_band'].astype('int')
data['HomeDistance_band']=data['HomeDistance_band'].astype('int')
# data_talent = data[data['SalaryHike_range']==4]
# data_normal = data[data['SalaryHike_range'].isin([1,2,3])]

data_talent = data.loc[(data['SalaryHike_range']==4) & (data['Age'] < 34)]

data_normal = data.loc[~(data['SalaryHike_range']==4) | ~(data['Age'] < 34)]
```

## 핵심인재그룹 추출

### 임금인상률과 근속년수

- 신입사원이라 임금인상률이 높은 것은 아닐까?에 대한 대답
- 신입사원이라 임금인상률이 높은 것은 아니다.

```python
#우선 신입사원이라 임금 상승률이 높은 것인지 확인
f, ax = plt.subplots(1,2, figsize=(20,10))

sns.distplot(data_talent['YearsAtCompany'],ax=ax[0])
sns.distplot(data_normal['YearsAtCompany'],ax=ax[1])

print("data_talent_근속년수 평균 :", data_talent['YearsAtCompany'].mean())
print("data_normal_근속년수 평균 :", data_normal['YearsAtCompany'].mean())
print("\n")
print(data_talent.groupby('Attrition')[['YearsAtCompany']].agg(['count', 'max', 'min', 'mean', 'median']))
print("\n")
print(data_normal.groupby('Attrition')[['YearsAtCompany']].agg(['count', 'max', 'min', 'mean', 'median']))
```

### 임금인상률과 이직 

- PercentSalaryHike 는 Attrition에 큰 영향을 미치지 않아 보임

```python
#임금상승률

sns.distplot(data['PercentSalaryHike'])
print(data.groupby('Attrition')[['PercentSalaryHike']].agg(['count', 'max', 'min', 'mean', 'median']))
sns.factorplot('Attrition', 'PercentSalaryHike', data=data) 
```

### 핵심인재그룹 추출

- 나이가 어리고, 임금인상률이 높은 그룹을 회사의 핵심인재 그룹으로 판단
- 이 그룹과 나머지 다른 그룹과 비교 
- 이 그룹은 다른 그룹에 비해 퇴사 결정에 있어 어떤 영향을 많이 받을까?

```python
data_talent = data.loc[(data['SalaryHike_range']==4) & (data['Age'] < 34)]

data_normal = data.loc[~(data['SalaryHike_range']==4) | ~(data['Age'] < 34)]
```

# EDA

## StockOptionLevel과 이직

### 핵심인재그룹

```python
f, ax = plt.subplots(1,1, figsize=(5,5))

sns.distplot(data_talent[data_talent['Attrition']=='Yes']['StockOptionLevel'])
sns.distplot(data_talent[data_talent['Attrition']=='No']['StockOptionLevel'])

print(data_talent.groupby('Attrition')[['StockOptionLevel']].agg(['count', 'max', 'min', 'mean', 'median']))
```

### 일반그룹

```python
sns.distplot(data_normal[data_normal['Attrition']=='Yes']['StockOptionLevel'])
sns.distplot(data_normal[data_normal['Attrition']=='No']['StockOptionLevel'])
print(data_normal.groupby('Attrition')[['StockOptionLevel']].agg(['count', 'max', 'min', 'mean', 'median']))
```

- 핵심인재 그룹에서는 퇴사 vs 안퇴사 간의 Stockoptionlevel mean 차이가 0.44
- 아닌 그룹에서는 mean 차이가 0.29
- 임금인상률이 높은 그룹에서 Stockoptionlevel 을 이직 이유로 더 고려함을 알 수 있다.
